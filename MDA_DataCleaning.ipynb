```python
# Rename id in df_project to facilitate merging
df_project_copy = df_project.copy()
df_project_copy.rename(columns={'id': 'projectID'}, inplace=True)
df_project_copy.head()
```

```python
agg_organizations = df_organization.groupby('projectID').agg(lambda x: '; '.join(x.dropna().astype(str))).reset_index()
agg_legalBasis = df_legalBasis.groupby('projectID').agg(lambda x: '; '.join(x.dropna().astype(str))).reset_index()
agg_publications = df_projectPublications.groupby('projectID').agg(lambda x: '; '.join(x.dropna().astype(str))).reset_index()

# 4. Merge everything into df_project_copy
merged_all = df_project_copy.copy()

merged_all = merged_all.merge(agg_organizations, on='projectID', how='left')
merged_all = merged_all.merge(agg_legalBasis, on='projectID', how='left')
merged_all = merged_all.merge(agg_publications, on='projectID', how='left')

# 5. Check result
print("Merged DataFrame shape:", merged_all.shape)
merged_all.head()
```

```python
print(merged_all.info())
```

```python
# merge other two dataframes 
# First merge: df_topics
merged_all = merged_all.merge(
    df_topics,
    on='projectID',
    how='left',
    suffixes=('', '_topics')  # prevent duplicate column names
)

# Then merge: df_webLink
merged_all = merged_all.merge(
    df_project_copy,
    on='projectID',
    how='left',
    suffixes=('', '_project')  # avoid duplicate suffix errors
)

# Check the final shape and preview
print("Final merged shape:", merged_all.shape)
merged_all.head()
```

```python
# Convert financial columns: replace comma with dot, then to numeric
for col in ['ecMaxContribution', 'totalCost']:
    df_minusUnnecessaryVar[col] = pd.to_numeric(
        df_minusUnnecessaryVar[col].astype(str).str.replace(',', '.', regex=False),
        errors='coerce'
    )
# Convert date columns to datetime
date_cols = ['startDate', 'endDate', 'ecSignatureDate']
for col in date_cols:
    df_minusUnnecessaryVar[col] = pd.to_datetime(df_minusUnnecessaryVar[col], errors='coerce')

# Convert publishedYear to string (safe if it contains semicolon-separated values)
if 'publishedYear' in df_minusUnnecessaryVar.columns:
    df_minusUnnecessaryVar['publishedYear'] = df_minusUnnecessaryVar['publishedYear'].astype(str)

# Check the updated data types
print(df_minusUnnecessaryVar.dtypes)
```

```python
# rename title_x to title
df_minusUnnecessaryVar.rename(columns={'title_x': 'title'}, inplace=True)
print(df_minusUnnecessaryVar.info())
```

```python
print(df_minusUnnecessaryVar['ecMaxContribution'].head(10))
print(df_minusUnnecessaryVar['ecContribution'].head(10))
print(df_minusUnnecessaryVar['netEcContribution'].head(10))
print(df_minusUnnecessaryVar['totalCost'].head(10))
```

```python
# Define a function to safely split and sum semicolon-separated values
def sum_contributions(val):
    if pd.isna(val):
        return 0.0
    try:
        numbers = [float(x.strip()) for x in val.split(';') if x.strip()]
        return sum(numbers)
    except:
        return 0.0

# Apply it to ecContribution and netEcContribution
df_minusUnnecessaryVar['ecContribution_sum'] = df_minusUnnecessaryVar['ecContribution'].apply(sum_contributions)
df_minusUnnecessaryVar['netEcContribution_sum'] = df_minusUnnecessaryVar['netEcContribution'].apply(sum_contributions)

# Now compare with ecMaxContribution
df_minusUnnecessaryVar['ec_match'] = np.isclose(
    df_minusUnnecessaryVar['ecContribution_sum'], 
    df_minusUnnecessaryVar['ecMaxContribution'], 
    atol=1e-2
)

# View mismatches
mismatches = df_minusUnnecessaryVar[~df_minusUnnecessaryVar['ec_match']]
print(mismatches[['ecMaxContribution', 'ecContribution_sum']])
```
